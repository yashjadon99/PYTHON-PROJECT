##### UNIMODEL DATA VISUALIZATION

# 

##### BIVARIATE VARIALE VISUALIZATION

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(), annot = True)
```

Observations:
1) By looking at the correlation matrix we can see that RM has a strong positive correlation with MEDV (0.7) where as LSAT has  a high negative correlation with MEDV(-0.74).

2) Based on the above observations we will RM and LSAT as our features. Using a scatter plot letâ€™s see how these features vary with MEDV.

```python

plt.figure(figsize=(20, 5))

a = ['RM', 'LSTAT']
b = df['MEDV']

for i, col in enumerate(a):
    plt.subplot(1, len(a) , i+1)
    x = df[col]
    y = b
    plt.scatter(x, y, marker='o', color = 'b', alpha = 0.5)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

Observations:
1) The prices increase as the value of RM increases linearly. There are few outliers and the data seems to be capped at 50.
2) The prices tend to decrease with an increase in LSTAT. Though it doesnâ€™t look to be following exactly a linear line.

# INSIGHTS

### 1) Distribution of Median value of owner-occupied homes

```python
plt.figure(figsize=(15,6))
plt.hist(df["MEDV"], bins=40, color = 'orange', edgecolor = 'white')
plt.xlabel("Median value of owner-occupied homes ($)")
plt.ylabel("Frequency")
plt.title("Distribution of MEDV")
plt.show()
```

##### The median value of owner-occupied homes (MEDV) ranges from around  5,000  to 50,000, with the majority of the values clustered around 20,000.

# 

### 2) Distribution of Average Number of Rooms

```python
plt.figure(figsize=(15,6))
plt.hist(df['RM'], bins=40, color='purple',edgecolor = 'white')
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Frequency')
plt.title('Distribution of Average Number of Rooms')
plt.show()
```

##### The average number of rooms per dwelling (RM) has a relatively normal distribution, with the majority of the values clustering around 6.

# 

### 3) Distribution of LSTAT

```python
plt.figure(figsize=(15,6))
plt.hist(df['LSTAT'], bins = 40, color='blue', edgecolor = 'white')
plt.title('Distribution of LSTAT')
plt.xlabel('Town index')
plt.ylabel('Percentage of lower status population')
plt.show()
```

##### The percentage of lower status population (LSTAT) has a right-skewed distribution, indicating that most of the towns have a relatively low percentage of lower status population.

# 

### 4) Average CRIM Rate vs RAD

```python
rad_group = df.groupby(['RAD']).agg({'CRIM': np.mean})
plt.plot(rad_group.index, rad_group['CRIM'])
plt.title('Average CRIM Rate vs RAD')
plt.xlabel('RAD')
plt.ylabel('Average CRIM Rate')
plt.show()
```

##### From the line graph, we can observe that there is a general trend of increasing mean CRIM rate as RAD increases

# 

### 5) Average NOX Rate vs RAD

```python
nox_group = df.groupby(['RAD']).agg({'NOX': np.mean})
plt.bar(nox_group.index, nox_group['NOX'], color='darkgreen')
plt.title('Average NOX Rate vs RAD')
plt.xlabel('RAD')
plt.ylabel('Average NOX Rate')
plt.show()
```

##### The bar plot shows that the average NOX rate is highest for RAD values of 24, 21 and 7, and lowest for RAD values of 8, 2, and 1. The plot also shows that there is a slight increasing trend in the average NOX rate with increasing RAD values.

# 

### 6) Average MEDV by CHAS

```python
chas_group = df.groupby(['CHAS']).agg({'MEDV': np.mean})
plt.figure(figsize=(10,5))
plt.bar(chas_group.index, chas_group['MEDV'], color=['skyblue', 'orange'], width= 0.5)
plt.title('Average MEDV by CHAS')
plt.xlabel('CHAS')
plt.ylabel('Average MEDV')
plt.xticks([0, 1], ['0', '1'])
plt.show()
```

##### From graph we can see that the MEDV: Median value of owner-occupied home is more where Charles River dummy variable is 1. Which means where ever Charles river is present, the MEDV is higher

## NOX Value and CRIM Rate

#### Making new column NOX_Category in which there are going to be 3 values - Low, Medium, High according to the NOX value

```python
bins = [0, 0.4, 0.6, 1]
df['NOX_Category'] = pd.cut(df['NOX'], bins, labels=['Low', 'Medium', 'High'])
```

```python
df
```

```python
counts = df['NOX_Category'].value_counts()
plt.figure(figsize = (20,12))
plt.bar(counts.index, counts.values, color = 'skyblue')
plt.title('Count of NOX Category')
plt.xlabel('NOX Category')
plt.ylabel('Count')
plt.show()
```

```python
counts
```

### 7) High NOX Value and CRIM Rate

```python
high_nox_crim = df[(df['NOX_Category'] == 'High') & (df['CRIM'] > 0)]

plt.figure(figsize=(25,10))
plt.bar(np.arange(len(high_nox_crim)), high_nox_crim['CRIM'], color = 'darkblue')
plt.xticks(np.arange(len(high_nox_crim)), high_nox_crim.index, rotation=90)
plt.xlabel('Towns')
plt.ylabel('CRIM Rate')
plt.title('High NOX Value and CRIM Rate')
plt.show()
```

### 8) Low NOX Value and CRIM Rate

```python
low_nox_crim = df[(df['NOX_Category'] == 'Low') & (df['CRIM'] > 0)]

plt.figure(figsize=(25,10))
plt.bar(np.arange(len(low_nox_crim)), low_nox_crim['CRIM'], color = 'purple')
plt.xticks(np.arange(len(low_nox_crim)), low_nox_crim.index, rotation=90)
plt.xlabel('Towns')
plt.ylabel('CRIM Rate')
plt.title('Low NOX Value and CRIM Rate')
plt.show()
```

### 9) INDUS variable distribution

```python
plt.figure(figsize = (20,10))
plt.hist(df['INDUS'], bins=50, edgecolor = 'white', color = 'red')
plt.xlabel('Proportion of non-retail business acres per town')
plt.ylabel('Frequency')
plt.title('Histogram of INDUS')
plt.show()
```

##### From the histogram of INDUS, we can see that the majority of towns have a proportion of non-retail business acres between 5% and 20%. There are relatively few towns with very low (less than 2.5%) or very high (more than 27.5%) values of INDUS.

# 

##### 10) Mean INDUS by RAD

```python
indus_by_rad = df.groupby('RAD')['INDUS'].mean()

plt.bar(indus_by_rad.index, indus_by_rad.values, color = 'blue')
plt.xlabel('RAD')
plt.ylabel('Mean INDUS')
plt.title('Mean INDUS by RAD')
plt.show()
```

##### The bar plot of INDUS by RAD shows the average value of INDUS for each value of RAD, which represents the index of accessibility to radial highways. We can observe that the average INDUS values are generally higher for lower values of RAD, indicating that areas with better accessibility to radial highways tend to have lower concentrations of non-retail businesses. However, there is some variability in INDUS values for each RAD value, suggesting that other factors may also influence the development patterns in these areas.







# 

##### 11) Heat map

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

OBSERVATION:

1) The strongest positive correlation is between RAD and TAX, which indicates that the areas with high accessibility to radial highways also tend to have higher property tax rates.

2) There is a strong negative correlation between DIS and both NOX and AGE. This suggests that the further away a property is from the employment centers, the lower the concentration of nitrogen oxides in the air and the newer the properties in the area.

3) There is a moderate positive correlation between RM and MEDV, which suggests that the median value of owner-occupied homes tends to increase with the average number of rooms per dwelling.

4) There is a moderate negative correlation between LSTAT and MEDV, which suggests that areas with a higher percentage of lower status of the population tend to have lower median home values.

5) There is also a moderate negative correlation between INDUS and DIS, which suggests that areas with a higher proportion of non-retail business acres tend to be further away from the employment centers.

6) There is a weak negative correlation between CRIM and both RM and MEDV, which indicates that areas with higher crime rates tend to have lower home values and fewer rooms per dwelling.

7) Overall, this heatmap can provide valuable insights into the relationships between different variables in the dataset, which can be helpful for making informed decisions related to real estate investments or policy-making.
